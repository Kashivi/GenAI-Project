```python
import torch
import torchaudio
import torch.nn as nn
import math
```

```python
# Question 1


# Create example tensor
x = torch.randn(32, 1, 64, 64)

# Flatten except batch dimension
x_flat = x.view(x.size(0), -1)

print("Flattened shape:", x_flat.shape)

# Input features for Linear layer
input_features = x_flat.shape[1]
print("Input features for Linear layer:", input_features)
```

```python
# Question 2

sample_rate = 16000
duration = 1

# Create dummy audio (1 second)
waveform = torch.randn(1, sample_rate * duration)

mel_transform = torchaudio.transforms.MelSpectrogram(
    sample_rate=16000,
    n_fft=400,
    hop_length=160,
    n_mels=64
)

mel_spec = mel_transform(waveform)

print("Mel Spectrogram shape:", mel_spec.shape)
```

```python
# Question 3

# Input tensor (batch_size=1)
x = torch.randn(1, 1, 64, 64)

conv = nn.Conv2d(1, 8, kernel_size=3, stride=1, padding=1)
pool = nn.MaxPool2d(kernel_size=2, stride=2)

out = conv(x)
out = pool(out)

print("Output shape:", out.shape)
```
